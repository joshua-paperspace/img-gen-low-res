```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "a Dali-esque painting of an astronaut riding a unicorn on mars"
image = pipe(prompt).images[0]

image.save("low_res.png")
```

```python
import requests
import torch
from PIL import Image
from io import BytesIO
print('0')
from diffusers import StableDiffusionImg2ImgPipeline

print('1')
device = "cuda"
# model_id_or_path = "runwayml/stable-diffusion-v1-5"
model_id_or_path = "stabilityai/stable-diffusion-2-1"
# model_id_or_path = 'nitrosocke/Nitro-Diffusion'
print('2')
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id_or_path, torch_dtype=torch.float16)
print('3')
pipe = pipe.to(device)

url = "https://raw.githubusercontent.com/CompVis/stable-diffusion/main/assets/stable-samples/img2img/sketch-mountains-input.jpg"

# response = requests.get(url)
# init_image = Image.open(BytesIO(response.content)).convert("RGB")
init_image = Image.open("low_res.png").convert("RGB")
init_image = init_image.resize((768, 512))

# prompt = "A fantasy landscape, trending on artstation"
prompt = "a Dali-esque painting of an astronaut riding a unicorn on mars"

images = pipe(prompt=prompt, image=init_image, strength=0.75, guidance_scale=7.5).images
images[0].save("high_res.png")
```
